```python
import pandas as pd

clean = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/earnings_text_clean.pkl').drop('text', axis=1)
```

```python
from gensim import corpora, models 
```

```python
dictionary = corpora.Dictionary(clean.clean.to_list())
```

```python
corpus = [dictionary.doc2bow(doc) for doc in clean.clean]
```

```python
num_topics = 10
earnings_lda = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=10)
```

```python
earnings_lda.print_topics()
```

```python
# save the model
earnings_lda.save('../data/earnings_lda.gensim')

# save dictionary
dictionary.save('../data/earnings_dictionary.gensim')

# corpus save
corpora.MmCorpus.serialize('../data/earnings_corpus.mm', corpus)
```
