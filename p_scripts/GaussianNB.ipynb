```python
import pandas as pd 
# load main data
X_train = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/X_train')
X_val = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/X_val')
X_test = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/X_test')

y_train = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/y_train')
y_val = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/y_val')
y_test = pd.read_pickle('/Volumes/MacAirSSD/Earnings/data/y_test')
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
```

```python
gnb = GaussianNB()
# train the model on the training set
gnb.fit(X_train, y_train)

# make prediction on the val data
y_val_pred = gnb.predict(X_val)

gnb_val_report = classification_report(y_val, y_val_pred)

print(f'Classification report on GNB validation: {gnb_val_report}')
```

```python
gnb_test_pred = gnb.predict(X_test)
gnb_test_report = classification_report(y_test, gnb_test_pred)
print(f'Classification report on GNB test')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import legacy
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
import numpy as np
y_train = np.array(y_train)
y_val = np.array(y_val)
y_test = np.array(y_test)
```

```python
# define the network model
model = Sequential()

model.add(Dense(32, input_dim = X_train_scaled.shape[1], activation = 'relu', )) # input layer
model.add(Dense(16, activation='relu')) # hidden layer
model.add(Dense(1, activation='sigmoid')) # Output layer for binary classification

# compile the model
model.compile(loss='binary_crossentropy', optimizer=legacy.Adam(learning_rate=0.01), metrics=['accuracy'])

# train the model
model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, validation_data = (X_val_scaled, y_val))

# evaluate on val set
val_loss, val_accuracy = model.evaluate(X_val_scaled, y_val)
print(f'Validation Accuracy: {val_accuracy}')

# after tuning hyper params on the val set
X_train_val_combined = np.concatenate([X_train_scaled, X_val_scaled], axis=0)
y_train_val_combined = np.concatenate([y_train, y_val], axis=0)

# combine train and validation, retrain and then evaluate on the testing set
model.fit(X_train_val_combined, y_train_val_combined, epochs=10, batch_size=32)
test_lost, test_accuracy = model.evaluate(X_test_scaled, y_test)

print(f'Testing Accuracy Score: {test_accuracy}')
```

```python
from sklearn.metrics import classification_report

# Assuming you've trained your neural network model and made predictions on the test set
y_test_pred = (model.predict(X_test_scaled) > 0.5).astype("int32").reshape(-1)

# Generate and print the classification report
report = classification_report(y_test, y_test_pred)
print(report)
```
